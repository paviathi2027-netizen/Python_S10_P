```python
import pandas as pd

reviews_df = pd.read_csv("/content/DA314_S1_Reviews_Data_Concept.csv")
reviews_df
```

```python
#create a series by extracting 'Rating' column and setting review_id as label
rating_series = pd.Series(reviews_df['rating'].values, index=reviews_df['restaurant_id'])
print(rating_series.head())
print(type(rating_series))
```

```python
#create a series by extracting 'review_text' column without lable
review_series = reviews_df['review_text']
print(review_series.head())
print(type(review_series))
```

```python
#create a series by extracting 'review_text' column without lable
review_series = reviews_df['review_text']
print(review_series.head())
print(type(review_series))
```

```python
#Subset using iloc (position)
fifth_review = review_series.iloc[4]
print(fifth_review)
```

```python
#Analysis the series
print("\nAnalysis of the rating series\n")
print(f"Mean: {rating_series.mean().round(1)}")
print(f"Median: {rating_series.median()}")
print(f"Mode: {rating_series.mode()[0]}")
print(f"Min: {rating_series.min()}")
print(f"Max: {rating_series.max()}")
```

```python
# Group by index (restaurant_id) and calculate average
average_ratings = rating_series.groupby(rating_series.index).mean()
average_ratings = average_ratings.round(1)
print("Average Rating By restaurant_id:")
print(rating_series.head())
```

```python
#Convert series to a datafreme
restaurant_ratings_df = average_ratings.to_frame(name="avg_rating")
# Rest label into column
restaurants_reviews_df = restaurant_ratings_df.reset_index("restaurant_id")
print(restaurant_ratings_df.head())
```

```python
import pandas as pd

# Read the restaurants file
restaurants_df = pd.read_csv("/content/DA314_S2_Restaurants_Data_Concept.csv")
print("Restaurants Data:")
print(restaurants_df.head())
```

```python

# Read the reviews file
reviews_df = pd.read_csv("/content/DA314_S2_Reviews_Data_Concept.csv")
print("\nReviews Data:")
print(reviews_df.head())
```

```python
#Join user with Order
orders_users = pd.merge(order_df,user_df, on='user_id', how='left')
#Join with Restaurants
orders_full_df = pd.merge(orders_users, restaurants_df,
                          on='restaurant_id', how='left',
                          suffixes=('_user','_restaurant')
                          )
orders_full_df.info()
```

```python
#Common column's of each table[orders_users, restaurants_df] ->Suffixes are added
orders_full_df[['name_user','name_restaurant']].head()
```

```python
restaurants_reviews_df = pd.merge(
    restaurants_df,
    reviews_df,
    on='restaurant_id',
    how='right',
    suffixes=('_restaurant', '_review')
)
print(restaurants_reviews_df.head())
```

```python
restaurants_reviews_df.isnull().sum()
```

```python
restaurants_reviews_df['rating_review'].fillna(
    restaurants_reviews_df['rating_review'].mean(),
    inplace=True
)

restaurants_reviews_df.isnull().sum()
```

```python
# check shape
restaurants_reviews_df.shape
```

```python
# drop rows where cuisine_type(Column(s)) is null value (NaN)
restaurants_reviews_df = restaurants_reviews_df.dropna(subset=['cuisine_type'])
```

```python
# drop rows with at least one NaN values(DROPNA)
restaurants_reviews_df.dropna(inplace=True)
```
